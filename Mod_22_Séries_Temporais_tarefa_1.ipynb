#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

**Calculando Amplitude:**

```python
# Selecionando as colunas que vejo como importantes de manter:

df1 = yf[["Adj Close", "High", "Low", "Volume"]]
df1.head()
```

```python
# Empilhando os dados por meio do stack(), para que eu consiga
# criar um df que me permita a análise, bem como a criação de uma
# nova coluna, que será a amplitude.

# Porém, para tanto, devo também resetar o índice, que estou fazendo
# via reset_index().

df1 = df1.stack()
df1 = df1.reset_index()

df1.head()
```

```python
df1.rename(columns={"Date":"date", "Ticker":"ticker",
                    "Adj Close":"adj close",
                    "High":"high",
                    "Low":"low",
                    "Volume":"volume"},
           inplace=True)



df1.head()
```

- Dúvida: Como altera a coluna "Price"? Ao meu ver, ela está na posição do índice, o que me fez querer alterar seu nome para index, porém não consegui, inclusive adotando o rename column, como fiz com as outras colunas.

```python
df1["amplitude"] = df1["high"] - df1["low"]
df1.head()
```

```python
fig = px.line(df1, x="date", y="amplitude", color="ticker", template="ggplot2")
fig.show()
```

A partir do gráfico, verifica-se que a NVIDIA - NVDA - apresentou a maior volatilidade ao longo do período, por mais que no início, antes de 22, estava abaixo quando comparada com o Pinterest - PINS -. A sua volatilidade e crescimento, porém, podem ser explicados justamente porque no período, as aplicações de IA's, em especial as generativas, passaram a serem difundidas, o que incrementou a demanda por processadores, que são extremamente necessários para o funcionamento de tais IA's.
