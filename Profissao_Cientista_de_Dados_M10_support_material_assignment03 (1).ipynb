* **Dúvida:**

A partir desse gráfico, eu pensei em segmentar o dataframe em dois momentos, um antes do ensino superior aumentar e o outro depois que ele aumentou. Por meio disso, iria querer calcular em média o tanto que o ensino secundário declinou em relação ao novo contexto, para verificar se o seu declínio se associa com o incremento, no mesmo instante, do contingente do ensino superior. Como eu poderia fazer isso em termos de código?

O que eu pensei, para tanto, estará nos códigos abaixo.

````python
# Filtrando o dataframe conforme o intervalo desejado:

df_edu.data_ref = pd.to_datetime(df_edu["data_ref"])

df_edu.dtypes
````

````python
dt_referencia = pd.to_datetime("2015-09-01")

# Filtrar o DataFrame para o intervalo desejado

intervalo1 = df_edu.loc[df_edu["data_ref"] < dt_referencia]
intervalo2 = df_edu.loc[df_edu["data_ref"] > dt_referencia]

x = intervalo1[intervalo1["educacao"] == "Secundário"].value_counts().sum()
x2 = intervalo1[intervalo1["educacao"] == "Superior completo"].value_counts().sum()

y = intervalo2[intervalo2["educacao"] == "Secundário"].value_counts().sum()
y2 = intervalo2[intervalo2["educacao"] == "Superior completo"].value_counts().sum()

print(f"Ensino Secundário antes da data de referência: {x}")
print(f"Ensino Superior Completo antes da data de referência: {x2}")
print("")
print(f"Ensino Secundário depois da data de referência: {y}")
print(f"Ensino Superior completo depois da data de referência: {y2}")
````

````python
# Calculando o percentual da variação do montante de estudantes
# no ensino secundário e superior:

pct1 = ((y/x) - 1).round(2)
pct2 = ((y2/x2) - 1).round(2)


print(f"Percentual da variação do ensino secundário: {pct1}")
print(f"Percentual da variação do ensino superior completo: {pct2}")
````
