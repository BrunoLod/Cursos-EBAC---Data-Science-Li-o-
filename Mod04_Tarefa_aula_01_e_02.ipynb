#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui

media_por_coluna = dados.mean(axis=1)
media_por_coluna
```

```python
dados["media_das_colunas"] = media_por_coluna
dados
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui

import math

dados["media_das_colunas"]
```

```python
# Como eu queria acessar cada valor da coluna que armazena a média
# das colunas, ao passo que fazer math.log(dados["media_das_colunas"]) não
# lograva êxito, decide criar um ciclo for para tanto. Com ele acessei cada
# elemento da coluna desejada, realizei o cálculo matemático, adicionando-o
# na criação de uma nova coluna no DataFrame.

for i in dados["media_das_colunas"]:

  log_med = math.log(i)

dados["log_med"] = log_med

dados
```
