#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# Descategorizando a coluna CODE_GENDER, para que ela
# possua em relação a cada genêro um número, sendo 1 para
# feminino e 0 para masculino, respectivamente.

df.sexo = df.sexo.map({"F": 1, "M": 0})

df.head()
````

````python
# Criando um novo dataframe com os dados "descategorizados", para que
# possam ser posteriormente trabalhados numa árvore de decisões.

df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
````

````python
# Alterando o caráter booleano das colunas adult_male e alone, para que sejam
# inteiros. Nesse caso, o valor 1 será para o True, enquanto que 0 para o False.

df_encoded.mau = df_encoded.mau.astype(int)

df_encoded.head()
````

````python
df_encoded.columns
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
# Fazer uma tabela com as variáveis possui_email e posse_de_veículo
# em relação a variável mau.

tabela_cruzada_email = pd.crosstab(df_encoded["possui_email"], df_encoded["mau"])

tabela_cruzada_email
````

````python
tabela_cruzada_veiculo = pd.crosstab(df_encoded["posse_de_veiculo_Y"], df_encoded["mau"])

tabela_cruzada_veiculo
````

````python
# Para veículo

prob_mau_veiculo_nao = tabela_cruzada_veiculo.loc[0, 1] / tabela_cruzada_veiculo[1].sum()

prob_mau_veiculo_sim = tabela_cruzada_veiculo.loc[1, 1] / tabela_cruzada_veiculo[0].sum()

# Para email

prob_mau_email_nao = tabela_cruzada_email.loc[0, 1] / tabela_cruzada_email[1].sum()

prob_mau_email_sim = tabela_cruzada_email.loc[1, 1] / tabela_cruzada_email[0].sum()
````

````python
print(f"A probabilidade de ser mau pagador e não ter veículo é de {round(prob_mau_veiculo_nao, 2)}")
print("")
print(f"A probabilidade de ser mau pagador e ter veículo é de {round(prob_mau_veiculo_sim, 2)}")
print("")
print(f"A probabilidade de ser mau pagador e não ter email é de {round(prob_mau_email_nao, 2)}")
print("")
print(f"A probabilidade de ser mau pagador e ter email é de {round(prob_mau_email_sim, 2)}")
````
