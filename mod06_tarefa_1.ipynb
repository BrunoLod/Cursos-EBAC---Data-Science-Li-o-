####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Como há muitas linhas nesse dataframe, acima de 25k, e eu tentei
# executar as células inicialmente com a quantidade citada, porém
# demarou muito, preferi retirar dos dados brutos uma amostra.

# Como o meu professor de estatística da faculdade (ESEG) diz que o nível
# de amostragem mínimo costuma ser, por consenso, maior ou igual a 30%,
# define uma proporção amostral de 30%, em termos decimais de 0.3.

# Proporção amostral
proporcao_amostragem = 0.3

# Realiza amostragem estratificada, armazenando numa nova varíavel.
# Estou agrupando o df a partir da coluna sexo, aplicando uma função
# anônima, lambda, para realizar a estratificação.

df_amostral = df.groupby('SEXO', group_keys=False).apply(lambda x: x.sample(frac=proporcao_amostragem))
df_amostral
```

```python
plt.rc("figure", figsize=(10,8))

# Criando uma plotagem que terá como base a criação de
# quatro gráficos, dois na porção de cima e os demais embaixo.

fig, axis = plt.subplots(2,2)

# Criando gráficos, por meio dos quais me permite saber
# a quantidade de cada variável:

sns.countplot(ax = axis[0,0], x="APGAR1", data=df_amostral)
sns.countplot(ax = axis[0,1], x="APGAR5", data=df_amostral)
sns.countplot(ax = axis[1,0], x="SEXO", data=df_amostral)
sns.countplot(ax = axis[1,1], x="GESTACAO", data=df_amostral)
sns.countplot(ax = axis[0,1], x="ESTCIVMAE", data=df_amostral)

# Inclinando os rótulos das colunas no segundo gráfico (axis[0, 1])
axis[1, 1].tick_params(axis='x', rotation=45)

plt.show()
```

```python
plt.rc("figure", figsize=(10,8))

# Criando uma plotagem que terá como base a criação de
# quatro gráficos, dois na porção de cima e os demais embaixo.

fig, axis = plt.subplots(2,2)

# Criando gráficos, por meio dos quais me permite saber
# a quantidade de cada variável:

sns.countplot(ax = axis[1,0], x="SEXO", data=df_amostral)
sns.countplot(ax = axis[1,1], x="GESTACAO", data=df_amostral)
sns.countplot(ax = axis[0,1], x="ESTCIVMAE", data=df_amostral)
sns.countplot(ax = axis[0,0], x="CONSULTAS", data=df_amostral)

# Inclinando os rótulos das colunas no segundo gráfico (axis[0, 1])
axis[1, 1].tick_params(axis='x', rotation=45)

plt.show()
```
